### Tabela Bronze de Produtos a Partir de Arquivo JSON

```python
# Imports
from pyspark.sql.functions import current_timestamp
```

```python
# Caminho para o arquivo
caminho = '/Volumes/workspace/ads_dados_brutos/arquivos/produtos.json'
```

```python
# Carrega o arquivo como um dataframe do Spark
df_ads = spark.read.json(caminho)
```

```python
# Visualiza os dados
#df_ads.show()
```

```python
# Cria uma coluna com a data que o registro foi recebido
novo_df = df_ads.withColumn('recebido_em', current_timestamp())
```

```python
# Visualiza os dados
#display(novo_df)
```

```python
# Define o banco para salvar a tabela
spark.sql("use database ads_db_01_inicial")
```

```python
# Grava a tabela com o formato delta
novo_df = novo_df.write.format("delta").mode("append").saveAsTable("ads_bronze_produtos")
```

```python
# Consulta a tabela
spark.sql('select * from ads_bronze_produtos limit 10').show()
```
