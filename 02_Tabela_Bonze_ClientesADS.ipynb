### Tabela Bronze de Clientes a Partir de Arquivo CSV

```python
# Imports
from pyspark.sql.functions import current_timestamp
```

```python
# Caminho para o arquivo
caminho = '/Volumes/workspace/ads_dados_brutos/arquivos/clientes.csv'
```

```python
# Carrega o arquivo como dataframe do Spark
df_ads = spark.read.csv(caminho, header = True, inferSchema = True)
```

```python
# Visualiza os dados
df_ads.show()
```

```python
# Adiciona uma coluna com a data que o registro foi recebido
novo_df = df_ads.withColumn('recebido_em', current_timestamp())
```

```python
# Visualiza os dados
display(novo_df)
```

```python
# Define o banco de dados (schema) a ser usado
spark.sql("use database ads_db_01_inicial")
```

```python
# Vamos usar arquitetura Delta Lake para gravar os dados no Data Lakehouse
# Isso permite usar propriedades ACID
novo_df = novo_df.write.format("delta").mode("append").saveAsTable("ads_bronze_clientes")
```

```python
# Consulta a tabela
spark.sql('select count(*) from ads_bronze_clientes').show()
```
